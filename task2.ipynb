```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd

nltk.download('vader_lexicon')

def perform_sentiment_analysis(sentence):
    sia = SentimentIntensityAnalyzer()
    compound_score = sia.polarity_scores(sentence)['compound']
    sentiment_label = 'Positive' if compound_score >= 0 else 'Negative'
    return compound_score, sentiment_label
data=pd.read_csv("senti.csv")
print(data)

user_input = input("\nEnter a sentence: ")
compound_score, sentiment_label = perform_sentiment_analysis(user_input)

print(f"Compound Score: {compound_score}")
print(f"Sentiment Analysis Label: {sentiment_label}")
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.bar(data['word'], [perform_sentiment_analysis(sentence)[0] for sentence in data['word']], color=['green' if rating == 1 else 'red' for rating in data['rating']])
ax.set_ylabel('Compound Score')
ax.set_title('Sentiment Analysis')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
